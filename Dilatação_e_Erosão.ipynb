##Dilatação

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_img(img):
  fig = plt.figure(figsize=(6,6))
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap='gray')
  plt.grid('on')
  plt.axis('off')
```

```python
img_original = np.zeros((6,6), dtype=np.uint8)
img_original
```

```python
coordenadas= [(1,3),(2,2),(3,2)]
for coord in coordenadas:
    img_original[coord]=1
display_img(img_original)
```

```python
kernel = np.array([[0,1,0],[0,1,0],[0,1,0]], dtype=np.uint8)
display_img(kernel)
```

```python
dilated_img = cv2.dilate(img_original,kernel,iterations=1)
display_img(dilated_img)
```

##Erosão

```python
img_original = np.zeros((6,6), dtype=np.uint8)
img_original
```

```python
coordenadas= [(1,4),(2,2),(2,3),(2,4),(3,2), (3,3), (3,4),(4,2),(4,3)]
for coord in coordenadas:
    img_original[coord]=1
display_img(img_original)
```

```python
erode_img = cv2.erode(img_original,kernel,iterations=1)
display_img(erode_img)
```
