```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()
```

```python
img = cv2.imread('/content/imagem_azul_P1.jpeg')

#Separa os canais
canais = cv2.split(img)
cores = ("b", "g", "r")
plt.figure()
plt.title("Histograma Colorido")
plt.xlabel("Intensidade")
plt.ylabel("NÃºmero de Pixels")
for (canal, cor) in zip(canais, cores):
  #Este loop executa 3 vezes, uma para cada canal
  hist = cv2.calcHist([canal], [0], None, [256], [0, 256])
  plt.plot(hist, color = cor)
  plt.xlim([0, 256])
plt.show()
```

```python
import cv2
import numpy as np

# Criar uma imagem em branco de 200x200 pixels com 3 canais de cor (RGB)
image = np.zeros((200, 200, 3), dtype=np.uint8)

# Definir a largura de cada faixa (aproximadamente 1/3 da largura total)
width = image.shape[1] // 3

# Desenhar a faixa azul
image[:, 0:width] = (255, 20, 20)  # Azul em BGR

# Desenhar a faixa verde
image[:, width:2*width] = (20, 255, 20)  # Verde em BGR

# Desenhar a faixa vermelha
image[:, 2*width:] = (20, 20, 255)  # Vermelho em BGR

# Gerar ruÃ­do branco (mesma dimensÃ£o da imagem e intensidade entre 0 e 50)
noise = np.random.randint(0, 120, (200, 200, 3), dtype=np.uint8)

# Adicionar o ruÃ­do branco Ã  imagem
noisy_image = cv2.add(image, noise)

# Exibir a imagem com ruÃ­do
cv2_imshow(noisy_image)
```

```python
#Separa os canais
canais = cv2.split(image)
cores = ("b", "g", "r")
plt.figure()
plt.title("Histograma Colorido")
plt.xlabel("Intensidade")
plt.ylabel("NÃºmero de Pixels")
for (canal, cor) in zip(canais, cores):
  #Este loop executa 3 vezes, uma para cada canal
  hist = cv2.calcHist([canal], [0], None, [256], [0, 256])
  plt.plot(hist, color = cor)
  plt.xlim([0, 256])
plt.show()
```
